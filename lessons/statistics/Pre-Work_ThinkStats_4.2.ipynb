## Exercise 4.2

**Exercise:** The numbers generated by `numpy.random.random` are supposed to be uniform between 0 and 1; that is, every value in the range should have the same probability.

Generate 1000 numbers from `numpy.random.random` and plot their PMF.  What goes wrong?

Now plot the CDF. Is the distribution uniform?

```python
# Generate 1000 numbers from numpy.random.random and create a PMF and CDF.

rand_1000=np.random.random(1000)
rand_1000_pmf=thinkstats2.Pmf(rand_1000)
rand_1000_cdf=thinkstats2.Cdf(rand_1000)
```

```python
# Plot a histogram of the PMF without modifying the width of displayed bars.
thinkplot.Hist(rand_1000_pmf)
thinkplot.Config(xlabel='Value', ylabel='Freq')

# The histogram seems to show nothing; however, this is due to the bars in the bar being too thin
# to be shown properly. Modifying the width of the bars in the histogram should help.
```

```python
# Plot a histogram of the PMF after modifying the width of displayed bars.
width=0.0005
thinkplot.Hist(rand_1000_pmf, width=width)
thinkplot.Config(xlabel='Value', ylabel='Freq')

# Now the histogram appears to be showing a uniform distribution.
```

```python
# Plot a step function of the PMF.
thinkplot.Pmf(rand_1000_pmf)
thinkplot.Config(xlabel='Value', ylabel='PMF')

# Using a step function to plot the PMF shows even more clearly the uniform distribution.
```

```python
# Plot the CDF.

thinkplot.Cdf(rand_1000_cdf)
thinkplot.Config(xlabel='Percentile rank', ylabel='CDF')

# As with all CDFs, the CDF is a straight line showing the distribution of percentile ranks is uniform.
```
